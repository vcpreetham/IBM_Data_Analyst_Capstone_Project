## Demo: How to list all tables


```python
# print all the tables names in the database
QUERY = """
SELECT name as Table_Name FROM
sqlite_master WHERE
type = 'table'
"""
# the read_sql_query runs the sql query and returns the data as a dataframe
pd.read_sql_query(QUERY,conn)
```

## Demo: How to run a group by query


```python
QUERY = """
SELECT Age,COUNT(*) as count
FROM master
group by age
order by age
"""
pd.read_sql_query(QUERY,conn)
```

# Hands-on Lab


## Visualizing distribution of data


### Histograms


Plot a histogram of `ConvertedComp.`


```python
# your code goes here
QUERY1 = """
SELECT ConvertedComp
FROM master
"""

df= pd.read_sql_query(QUERY1, conn)
df.hist( bins=30, figsize=(5,5) )
```

### Box Plots


Plot a box plot of `Age.`


```python
# your code goes here
QUERY2 = """
SELECT Age
FROM master
"""

Age_df= pd.read_sql_query(QUERY2, conn)
Age_df.plot(kind='box', figsize=(4,10))
```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
# your code goes here
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

QUERY3 = """
SELECT Age, WorkWeekHrs
FROM master
"""

df_scatter = pd.read_sql_query(QUERY3,conn)
df_scatter.plot(
    kind = 'scatter',
    x='Age',
    y='WorkWeekHrs',
)
plt.title('scatter plot of Age and WorkWeekHrs')
plt.xlabel('Age')
plt.ylabel('Number of Work Week Hours')

plt.show()
```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
# your code goes here

QUERY4 = """
SELECT Age, WorkWeekHrs, CodeRevHrs
FROM master
"""

df_bub=pd.read_sql_query(QUERY4,conn)
sns.set_context("talk", font_scale=1.1)
plt.figure(figsize=(10,6))
sns.scatterplot(data=df_bub, x='WorkWeekHrs', y='CodeRevHrs', size='Age', hue='Age', alpha=0.6, sizes=(10, 500)) 

plt.title('WorkWeekHrs and CodeRevHrs By Age', size=14) 
plt.xlabel('WorkWeekHrs', size=12) 
plt.ylabel('CodeRevHrs', size=12)

plt.show()
```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
# your code goes here
QUERY5 = """ 
SELECT DatabaseDesireNextYear, COUNT(*) as count 
from DatabaseDesireNextYear 
group by DatabaseDesireNextYear 
order by count(DatabaseDesireNextYear) DESC LIMIT 5 
""" 

pie_df=pd.read_sql_query(QUERY5,conn)  
pie_df.set_index('DatabaseDesireNextYear', inplace=True) 

colors_list=['gold', 'lightgreen', 'lightskyblue', 'pink', 'lightcoral']

pie_df['count'].plot(kind='pie', figsize=(20,6), autopct='%1.1f%%', labels=None, startangle=90, colors=colors_list, shadow=True, pctdistance=1.12)  

plt.legend(labels=pie_df.index, loc='upper right')
plt.title('Top 5 Databases Respondents Wish To Learn') 
plt.axis('equal')
plt.show()
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
# your code goes here
QUERY6 = """
SELECT WorkWeekHrs, CodeRevHrs, Age FROM master
WHERE Age BETWEEN 30 AND 35
"""
stack_df = pd.read_sql_query(QUERY6,conn)
df1 = stack_df.groupby('Age').median()

df1.plot(kind='bar', figsize=(10, 6), stacked=True)

plt.title('Stacked Bar Chart of Median WorkWeekHrs and CodeRevHrs for Those Age 30 to 35')
plt.legend(bbox_to_anchor = (1.4, 0.6), loc='center right')
plt.show()
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
# your code goes here
QUERY7 = """
SELECT ConvertedComp, Age FROM master
WHERE Age BETWEEN 45 AND 60
"""
line_df = pd.read_sql_query(QUERY7,conn)
df1 = line_df.groupby('Age').median()

df1.plot(kind='line', figsize=(10, 6))

plt.title('Line Chart of ConvertedComp for all ages from 45 to 60')
plt.ylabel('ConvertedComp')
plt.show()
```

### Bar Chart


Create a horizontal bar chart using column `MainBranch.`


```python
# your code goes here
QUERY7 = """
SELECT MainBranch, COUNT(*) AS MainBranch 
FROM master
GROUP BY MainBranch
"""
bar_df = pd.read_sql_query(QUERY7,conn)

bar_df.plot(kind='barh', figsize=(10, 6), color='green')

plt.title('horizontal bar chart using column MainBranch')
plt.show()
```
